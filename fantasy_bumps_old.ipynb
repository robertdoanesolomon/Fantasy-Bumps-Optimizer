```python
import pandas as pd

import numpy as np
from itertools import combinations_with_replacement


    
```

```python
# Load the CSV file
file_path = './Fantasy bumps - Sheet1.csv'  # Update this to the correct path if needed
data = pd.read_csv(file_path)

# Extract the relevant columns
crews = data['crew']
costs = data['cost'].values
returns = data['return'].values

# Generate all possible combinations of picking up to 9 crews with repetition
all_combinations = np.array(list(combinations_with_replacement(range(len(crews)), 9)))

# Vectorized calculation of costs and returns
combination_costs = costs[all_combinations].sum(axis=1)
combination_returns = returns[all_combinations].sum(axis=1)

print("HAVE YOU UPDATED THE AMOUNT OF CRABS YOU HAVE NOW?")

# Filter valid combinations (cost <= 1000)
valid_mask = combination_costs <= 2199
valid_combinations = all_combinations[valid_mask]
valid_returns = combination_returns[valid_mask]
valid_costs = combination_costs[valid_mask]

# Get indices of top 10 combinations
top_indices = np.argsort(valid_returns)[-10:][::-1]

# Display the top 10 combinations
for idx, top_idx in enumerate(top_indices):
    combination = valid_combinations[top_idx]
    best_crews = crews[combination].values
    best_total_cost = valid_costs[top_idx]
    total_return = valid_returns[top_idx]
    
    best_combination_details = pd.DataFrame({
        'crew': best_crews,
        'cost': costs[combination],
        'return': returns[combination]
    })

    print(f"Combination {idx + 1} with the Highest Return:")
    print(best_combination_details)
    print(f"Total Cost: {best_total_cost}")
    print(f"Total Return: {total_return}")
    print("\n")
```
